```python
ran = range(1,608)
```

```python
# create empty array of html_docs 
# for each i in range save soup of html doc in arr of html_docs
# for each html_doc inside arr of html_docs, parse elements.text from class menukad
# save all elements.texts in words arr
# clean nekudot in words arr
```

```python
html_docs = []

for i in ran:
    r = requests.get(f"https://www.pealim.com/dict/?page={i}")
    html_doc  = r.text
    html_docs.append(html_doc)
```

```python
soups = []

for html_doc in html_docs:
    soup = BeautifulSoup(html_doc, 'html.parser')
    soups.append(soup)
```

```python
words = []
for soup in soups:
    all_elements = soup.find_all(class_="menukad")
    for e in all_elements:
        e = e.text
        words.append(e)
```

```python
import re

def remove_nekudot(text):
    # Use a regular expression to remove all Hebrew vowel points
    cleaned_text = re.sub(r'[\u0591-\u05BD\u05BF-\u05C2\u05C4-\u05C7\u05F3\u05F4]', '', text)
    return cleaned_text
```

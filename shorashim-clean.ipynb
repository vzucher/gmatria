## Creating Combinations Dictionary

```python
columns_dict = {letter: [] for letter in alefbet}

combinations_2 = list(product(alefbet, repeat=2))
combinations_2_strings = [''.join(combo) for combo in combinations_2]
df_comb2 = pd.DataFrame(combinations_2_strings, columns=["Combinations"])

combinations_3 = list(product(alefbet, repeat=3))
combinations_3_strings = [''.join(combo) for combo in combinations_3]
df_comb3 = pd.DataFrame(combinations_3_strings, columns=["Combinations"])

for combo in df_comb2['Combinations']:
    starting_letter = combo[0]
    columns_dict[starting_letter].append(combo)
    
for combo in df_comb3['Combinations']:
    starting_letter = combo[0]
    columns_dict[starting_letter].append(combo)
    
df_final = pd.DataFrame(columns_dict)
df_final = df_final.fillna('')
df_final.to_csv("shorashim-blisofit.csv")    
```

```python
sofit_dict = {
    "כ": "ך",
    "מ": "ם",
    "נ": "ן",
    "פ": "ף",
    "צ": "ץ"
}
```

```python
def replace_sofit(string):
    
    sofit_dict = {
    "כ": "ך",
    "מ": "ם",
    "נ": "ן",
    "פ": "ף",
    "צ": "ץ"
    }
    # Check if the last character is one of the normal letters
    last_char = string[-1]
    if last_char in sofit_dict:
        # Replace with the corresponding sofit letter
        replaced_string = string[:-1] + sofit_dict[last_char]
        return replaced_string
    else:
        return string

# Test the function
string1 = "אבכ"
string2 = "יכפ"

replaced_string1 = replace_sofit(string1)
replaced_string2 = replace_sofit(string2)

print(replaced_string1)  # Output: "אבך"
print(replaced_string2)  # Output: "יכף"
```

```python
import re

def remove_nekudot(text):
    # Use a regular expression to remove all Hebrew vowel points
    cleaned_text = re.sub(r'[\u0591-\u05BD\u05BF-\u05C2\u05C4-\u05C7\u05F3\u05F4]', '', text)
    return cleaned_text
```

## Measuring Similarity through distance calc

Calculating semantic similarity using pre-trained word embeddings like Word2Vec, FastText, or BERT embeddings is typically done using techniques such as cosine similarity or Euclidean distance, rather than K-means or K-nearest neighbors (KNN). Here's a brief explanation of how each of these techniques is commonly used for semantic similarity:

1. **Cosine Similarity**: This is a common method for measuring the similarity between two vectors, such as word embeddings. Given two word embeddings, you can calculate the cosine similarity between them. The cosine similarity ranges from -1 to 1, with higher values indicating greater similarity. It is often used for tasks like finding the similarity between word vectors or document vectors.

2. **Euclidean Distance**: Similar to cosine similarity, you can also calculate the Euclidean distance between two word embeddings. However, while cosine similarity measures the cosine of the angle between two vectors, the Euclidean distance measures the actual distance between the points in the multi-dimensional space. Smaller Euclidean distances indicate greater similarity.

3. **K-means**: K-means is a clustering algorithm, not typically used for measuring semantic similarity directly. It is used to group similar data points into clusters. In the context of semantic similarity, you might use K-means to cluster words or phrases based on their embeddings but not to directly measure similarity between two specific words or phrases.

4. **K-nearest neighbors (KNN)**: KNN is a classification or regression algorithm that finds the K-nearest data points to a query point. It can be used for tasks like finding the nearest neighbors of a word or phrase based on their embeddings, which indirectly measures similarity. However, KNN itself is not a direct measure of semantic similarity but a way to find similar data points in a dataset.

To calculate semantic similarity between two words or phrases using pre-trained word embeddings, you would typically compute the cosine similarity or Euclidean distance between the embeddings of those words or phrases. K-means and KNN might come into play when you're working on tasks like clustering or finding nearest neighbors of words but are not the primary methods for measuring similarity.

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Autocomplete:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

    def search(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return []
            node = node.children[char]
        return self._find_words_with_prefix(node, prefix)

    def _find_words_with_prefix(self, node, prefix):
        suggestions = []
        if node.is_end_of_word:
            suggestions.append(prefix)
        for char, child_node in node.children.items():
            suggestions.extend(self._find_words_with_prefix(child_node, prefix + char))
        return suggestions

# Usage example:
autocomplete = Autocomplete()
words = ["apple", "appetizer", "banana", "bat", "cat"]
for word in words:
    autocomplete.insert(word)

user_input = "app"
suggestions = autocomplete.search(user_input)
print(suggestions)  # Output: ['appetizer', 'apple']
```

## Creating Combinations Dictionary

```python
# Generate all possible combinations of 2 and 3 letters from hebrew_alphabet
combinations_2 = list(product(alefbet, repeat=2))
combinations_3 = list(product(alefbet, repeat=3))

# Concatenate each sequence of 3 letters into one string
combinations_2_strings = [''.join(combo) for combo in combinations_3]
combinations_3_strings = [''.join(combo) for combo in combinations_3]


# Create a DataFrame
df = pd.DataFrame(combinations_3_strings, columns=["Combinations"])

# Create a dictionary to store the columns
columns_dict = {letter: [] for letter in alefbet}

# Populate the dictionary with sequences
for combo in df['Combinations']:
    starting_letter = combo[0]
    columns_dict[starting_letter].append(combo)

# Create a DataFrame from the dictionary
df_final = pd.DataFrame(columns_dict)

# Fill missing values with an empty string
df_final = df_final.fillna('')

# Print the DataFrame
df_final.to_csv("shorashim-blisofit.csv")
```

## Cleaning Sofits

```python
# map & access all values in Dataframe
# for each value in Dataframe is a string, access last char of string
# substitute sofits
```

```python
# replace sofits function
```

```python
sofits = ["ך", "ם", "ן", "ף", "ץ"]
```

```python
regular = ["כ", "מ", "נ", "פ", "צ"]
```

```python
sofit_dict = {
    "כ": "ך",
    "מ": "ם",
    "נ": "ן",
    "פ": "ף",
    "צ": "ץ"
}
```

```python
def replace_sofit(string):
    
    sofit_dict = {
    "כ": "ך",
    "מ": "ם",
    "נ": "ן",
    "פ": "ף",
    "צ": "ץ"
    }
    # Check if the last character is one of the normal letters
    last_char = string[-1]
    if last_char in sofit_dict:
        # Replace with the corresponding sofit letter
        replaced_string = string[:-1] + sofit_dict[last_char]
        return replaced_string
    else:
        return string

# Test the function
string1 = "אבכ"
string2 = "יכפ"

replaced_string1 = replace_sofit(string1)
replaced_string2 = replace_sofit(string2)

print(replaced_string1)  # Output: "אבך"
print(replaced_string2)  # Output: "יכף"
```

```python
df_final = df_final.applymap(replace_sofit)
df_final.to_csv("shorashim.csv")
df_final
```
